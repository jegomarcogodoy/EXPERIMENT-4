```python
import pandas as pd
```

```python
df = pd.read_csv('board2.csv')
df
```

```python
# For 1a.
# Set constants
constant_H = 'Luzon'
constant_T = 'Instrumentation'
# Create the data frames
Instru = df.loc[(df['Electronics']>70)&(df['Track']==constant_T)&(df['Hometown']==constant_H)][['Name', 'GEAS', 'Electronics']]
Instru
```

```python
# For 1b.
# Set the constants
const_H = 'Mindanao'
const_G = 'Female'
# Create the "Average" column
df['Average'] = df[['Electronics', 'GEAS', 'Math', 'Communication']].mean(axis=1)
# Filter the DataFrame
Mindy = df.loc[(df['Average'] >= 55) &(df['Hometown'] == const_H) &(df['Gender'] == const_G),['Name', 'Track', 'Electronics']]
Mindy
```

```python
# For 2
import matplotlib.pyplot as plt
```

```python
# Create a bar graph for each feature
fig, axes = plt.subplots(1, 3, figsize=(20, 4), sharey=True)

# For Gender
axes[0].bar(df['Gender'], df['Average'], color='pink')
axes[0].set_title('Average by Gender')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Average')

# For Track
axes[1].bar(df['Track'], df['Average'], color='purple')
axes[1].set_title('Average by Track')
axes[1].set_xlabel('Track')

# For Hometown
axes[2].bar(df['Hometown'], df['Average'], color='teal')
axes[2].set_title('Average by Hometown')
axes[2].set_xlabel('Hometown')

plt.show()
```

To answer the question: Does chosen track in college, gender, or hometown contributes to a higher average score?

Yes, a better average score is a result of all of them. The displayed bar graphs demonstrate that women achieve a higher average in terms of gender. Students who selected for the microelectronic track had a higher average in terms of track. Lastly, Luzon residents have a higher average when it comes to hometown.
